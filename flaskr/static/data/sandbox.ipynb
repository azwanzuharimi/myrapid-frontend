```python
df1=df.loc[0:4901]
df1=df1.rename(columns={df1.columns[0]:'date'})
df1
```

```python
df2=df.loc[4903:9805]
df2.columns=df2.loc[4903]
df2=df2.loc[4904:]
df2=df2.rename(columns={df2.columns[0]:'date'})
df2=df2.reset_index(drop='index')
df2
```

```python
df3=df.loc[9807:]
df3.columns=df3.loc[9807]
df3=df3.loc[9808:]
df3=df3.rename(columns={df3.columns[0]:'date'})
df3=df3.reset_index(drop='index')
df3
```

```python
df=pd.merge(df1,df2,how='left',on='date')
df=pd.merge(df,df3,how='left',on='date')
df.columns=pd.DataFrame(df.columns.to_list())[0].str.lower()
df
```

```python
df.to_parquet('exchange-rates.parquet')
```
